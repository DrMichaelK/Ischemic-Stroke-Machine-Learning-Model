```python
import os
import zipfile

def extract_zip_files(directory):
    # Walk through the directory tree
    for root, dirs, files in os.walk(directory):
        # Loop through all files in the current directory
        for file in files:
            # Check if the file is a ZIP file
            if file.endswith('.zip'):
                zip_path = os.path.join(root, file)

                # Set extraction path to the same location as the ZIP file
                extract_path = root

                try:
                    # Try extracting the ZIP file
                    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                        # Get the list of files in the ZIP archive
                        file_list = zip_ref.namelist()

                        if not file_list:
                            print(f"{file} is empty and has no files to extract.")
                            continue

                        # Extract files
                        zip_ref.extractall(extract_path)
                        print(f"Extracted {file} in {extract_path}")

                        # Check and log extracted files
                        extracted_files = [f for f in file_list if os.path.isfile(os.path.join(extract_path, f))]
                        if extracted_files:
                            print(f"Extracted files from {file}: {extracted_files}")
                        else:
                            print(f"Warning: {file} extracted but no files found in the target directory (could be empty folders).")

                except zipfile.BadZipFile:
                    print(f"Failed to extract {file}. The file may be corrupted.")
                except Exception as e:
                    print(f"An error occurred while extracting {file}: {e}")
# Replace the directory path with the path to your Google Drive folder
extract_zip_files("/content/drive/MyDrive/Infarcts/Lacunar")
```

```python
import os
from pdf2image import convert_from_path
import pytesseract
shared_folder_path = '/content/drive/MyDrive/Infarcts/Lacunar'

for root, dirs, files in os.walk(shared_folder_path):
    for file in files:
        if file.lower().endswith('.pdf'):
            pdf_path = os.path.join(root, file)
            txt_filename = os.path.splitext(file)[0] + '.txt'
            txt_path = os.path.join(root, txt_filename)
            # Convert PDF pages to images
            try:
                images = convert_from_path(pdf_path)
            except Exception as e:
                print(f"Error converting {pdf_path} to images: {e}")
                continue
            # Use OCR to extract text from each image (page)
            ocr_text = ''
            for img in images:
                try:
                    ocr_text += pytesseract.image_to_string(img)
                except Exception as e:
                    print(f"Error extracting text from image in {pdf_path}: {e}")
                    continue
            # Save the extracted text to a .txt file
            try:
                with open(txt_path, 'w', encoding='utf-8') as txt_file:
                    txt_file.write(ocr_text)
                print(f"Extracted text saved to {txt_path}")
            except Exception as e:
                print(f"Error saving text to {txt_path}: {e}")
```
